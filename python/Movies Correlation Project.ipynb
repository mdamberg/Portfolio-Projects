## Change data type of columns Budget and Gross to int

```python

#because we have some NaN values, we need to tell python how to handle those first b4 converting to int

df['budget'] = pd.to_numeric(df['budget'], errors = 'coerce')

df['gross'].fillna(0, inplace=True)

df['votes'].fillna(0, inplace=True)


# replace NaN values with default value of 0

df['budget'].fillna(0, inplace=True)

df['gross'].fillna(0, inplace=True)

df['votes'].fillna(0, inplace=True)


# Now we can cast the columns as int
    
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

### Numerizing non numerical columns
- Company and other cols in df are not numeric and do not get included in df.corr()

-  To work around this we can create a numeric represenation of this column to include it 

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```
